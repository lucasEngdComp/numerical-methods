```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
```

```python
def arctangle(x,n):
  return ((-1)**n*(x)**(2*n+1))/np.math.factorial(2*n+1)
```

PRIMEIRO MÉTODO - MARCHIN 

b) Fórmula de Machin

$$\frac {π}{4} = 4 arctan\frac {1}{5} - arctan\frac{1}{239}$$

```python
machin = 4*np.arctan(1/5) - np.arctan(1/239)
machin_index = 0
lista_machin = []

for k in range(21):
  machin_index = 4 * (4 *  arctangle((1/5), k) - arctangle((1/239), k))
  lista_machin.append(machin_index)
print(lista_machin)
```

```python
#listainv = lista_machin[::-1]
```

```python
x = np.arange(0,20)
lista_machin = np.array(lista_machin)

erro_machin = np.abs(np.diff(lista_machin) / lista_machin[0:20])
plt.figure(figsize = (12, 8))
plt.plot(x, erro_machin, "black", label = "machin")
plt.title("Fórmula de machin")
plt.xticks(x, x)
plt.grid(ls=":")
plt.xlabel("$Termos$")
plt.ylabel("$f(x)$")
plt.legend()
```

SEGUNDO MÉTODO - HUTTON 

b) Fórmula de Hutton

$$\frac {π}{4} = 4 arctan\frac {1}{2} - arctan\frac{1}{3}$$

```python
hutton = 4*np.arctan(1/2) + np.arctan(1/3)
hutt_list = []
hut_index = 0

for k in range(21):
  hut_index = 4 * (arctangle(1/2, k) + arctangle(1/3, k))
  hutt_list.append(hut_index)

print(hutt_list)
```

```python
hutt_list = np.array(hutt_list)
hutterror = np.abs(np.diff(hutt_list)/hutt_list[0:20])
plt.figure(figsize=(12,8))
plt.plot(x, hutterror, "cyan", label = "Hutton")
plt.title("Fórmula de Hutton")
plt.xticks(x, x)
plt.grid(ls=":")
plt.xlabel("$Termos$")
plt.ylabel("$f(x)$")
plt.legend()
```

TERCEIRO MÉTODO - CLAUSEN 

c) Fórmula de clausen

$$\frac {π}{4} = 2arctan\frac {1}{3} - arctan\frac{1}{7}$$

```python
clausen_formula = 2*np.arctan(1/3) + np.arctan(1/7)
clausen_list = []
clausen_index = 0

for u in range(21):
  clausen_index = 4 * (2 * arctangle(1/3, u) + arctangle(1/7, u))
  clausen_list.append(clausen_index)
print(clausen_list)
```

```python
clausen_list = np.array(clausen_list)
error_clausen = np.abs(np.diff(clausen_list)/clausen_list[0:20])
plt.figure(figsize = (12, 8))
plt.plot(x, error_clausen, "red", label = "Clausen")
plt.title("Fórmula de Clausen")
plt.xticks(x, x)
plt.grid(ls=":")
plt.xlabel("$Termos$")
plt.ylabel("$f(x)$")
plt.legend()
```
